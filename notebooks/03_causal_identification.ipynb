# Notebook 3: Causal Identification Strategy

## Overview

This notebook implements the causal identification strategy for estimating the effect of tertiary education expansion on youth unemployment.

**Research Question**: Does tertiary education expansion causally reduce youth unemployment in MENA countries?

**Identification Approach**: 
- Define treatment and control groups based on median-split of tertiary enrollment
- Verify covariate balance between groups
- Prepare data for Difference-in-Differences and Propensity Score Matching estimation

The core challenge is estimating the counterfactual - what would have happened to treated countries without education expansion.


```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../src')
from data_processing import load_panel_data, create_treatment_indicator, check_balance
from visualization import plot_balance
import warnings
warnings.filterwarnings('ignore')

df = load_panel_data('../data/processed/mena_panel_clean.csv')

# Create treatment groups (median split)
df_treated, treatment_info = create_treatment_indicator(df, 'tertiary_enrollment', None, 'median_year')

print(f"Treated countries: {treatment_info['n_treated']}")
print(f"Control countries: {treatment_info['n_control']}")
print(f"Treatment period: {treatment_info['post_period']}")

# Check balance
pre_data = df_treated[df_treated['post'] == 0]
covariates = ['gdp_per_capita', 'population_growth', 'education_spending', 'urban_population']
balance = check_balance(pre_data, covariates, 'treated')
display(balance)

# Save
df_treated.to_csv('../data/processed/mena_panel_with_treatment.csv', index=False)
print("\n✓ Treatment groups defined and saved")
```
